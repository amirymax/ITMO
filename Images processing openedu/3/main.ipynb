# Task 1

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('3_1.jpg')
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Построение гистограммы
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Определение максимального значения
max_value = max(histogram)

# Вывод максимального значения
print(f"Максимальное значение функции h(n): {max_value}")

# Построение гистограммы с использованием Matplotlib
plt.plot(histogram)
plt.title('Гистограмма изображения')
plt.xlabel('Яркость пикселей')
plt.ylabel('Количество пикселей')
plt.show()
```

```python
max_index = int(histogram.argmax())

# Определение интенсивности n при максимальном значении
max_intensity = max_index

# Вывод интенсивности n
print(f"Интенсивность n при максимуме h(n): {max_intensity}")
```

# Task 3

```python
# Загрузка изображения в формате BGR (цветовая модель OpenCV)
image_path = "3_3.jpg"
```

```python
bgr_image = cv2.imread(image_path)

# Разделение изображения на каналы B, G и R
b_channel, g_channel, r_channel = cv2.split(bgr_image)

# Линейная нормировка по каждому каналу
b_channel_normalized = cv2.normalize(b_channel, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
g_channel_normalized = cv2.normalize(g_channel, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
r_channel_normalized = cv2.normalize(r_channel, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

# Соединение каналов обратно в изображение
normalized_image = cv2.merge([b_channel_normalized, g_channel_normalized, r_channel_normalized])
b, g, r = normalized_image[437, 671]
print(r,g,b)
# Отображение изображения
# cv2.imshow("Normalized Image", normalized_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
image_path = "3_5.jpg"
```

```python
bgr_image = cv2.imread(image_path)

# Преобразование в палитру RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# Отображение исходного изображения
plt.figure(figsize=(6, 6))
plt.imshow(rgb_image)
plt.title("Исходное изображение")
plt.axis("off")
plt.show()

# Извлечение двузначного числа из верхнего левого угла
number_roi = rgb_image[:50, :50, :]

# Показ извлеченного двузначного числа
plt.figure(figsize=(2, 2))
plt.imshow(number_roi)
plt.title("Извлеченное число")
plt.axis("off")
plt.show()

# Применение bilateralFilter с различными значениями параметра d
d_values = [5, 9, 15]

for d_value in d_values:
    # Применение bilateralFilter
    filtered_image = cv2.bilateralFilter(rgb_image, d=d_value, sigmaColor=75, sigmaSpace=75)

    # Извлечение двузначного числа после применения bilateralFilter
    filtered_number_roi = filtered_image[:50, :50, :]

    # Показ извлеченного числа после bilateralFilter
    plt.figure(figsize=(2, 2))
    plt.imshow(filtered_number_roi)
    plt.title(f"Parameter d = {d_value}")
    plt.axis("off")
    plt.show()
```
