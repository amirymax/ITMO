# Task 1


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '4_1.jpg'
```

```python
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Выполнение дискретного преобразования Фурье
f_transform = np.fft.fft2(img)
print(f_transform[521, 1151])
```

```python
fshift = np.fft.fftshift(f_transform)
fshift[439][640] == f_transform[0][0]

# fshift.shape
```

```python
magnitude_spectrum = np.abs(fshift)
# plt.imshow(magnitude_spectrum, cmap='gray')
print('min:', np.min(magnitude_spectrum), 'max:', np.max(magnitude_spectrum))
```

```python
magnitude_spectrum = np.log(1+np.abs(fshift))
minn = np.min(magnitude_spectrum)
maxx = np.max(magnitude_spectrum)
print(maxx - minn)
```

```python
fshift[439 - 80 : 439 + 80, 640 - 80 : 640 + 80] = 0
magnitude_spectrum = np.log(1 + np.abs(fshift))
plt.imshow(magnitude_spectrum, cmap = 'gray')
```

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
plt.imshow(img_back, cmap = 'gray')
img_back[661][441]
```

```python
image_path = '4_2.jpg'
```

```python
img = cv2.imread(image_path)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
laplacian = cv2.Laplacian(img, -1)
plt.imshow(laplacian, cmap='gray')
# print('Сумма интенсивностей:', np.sum(laplacian))
b, g, r = cv2.split(laplacian)
print(np.sum(b) + np.sum(g) + np.sum(r))
```

```python
canny = cv2.Canny(img, 100, 200)
plt.imshow(canny, cmap='gray')
np.sum(canny)
```
